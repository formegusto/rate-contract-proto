```python
now_month = 12
rate_type="종합계약"

DB = KETIDB()
DB.connect()
datas = DB.find()

ds = DPPSUPPORTER(datas=datas, now_month=now_month)
households = ds.get_households(rate_type=rate_type)

PUBLIC_METER = 50000
exam_apt_meter = ds.sum_df.sum(axis=1)['kwh'] + PUBLIC_METER
    
mgmt = MGMTOFFICE(rate_type=rate_type, 
                  APT_METER=exam_apt_meter, 
                  now_month=now_month,
                  households=households,
                  peak_df=ds.peak_df)

IPython.display.clear_output()

mgmt.bill()
```

```python
PERCENTAGE_STEP = 5
LAST_STEP = 90
HOUSEHOLD_RATE = ds.sum_df.sum(axis=1)['kwh'] 
HOUSEHOLD_RATE_100 = HOUSEHOLD_RATE * 100

for PUBLIC_PERCENTAGE in range(PERCENTAGE_STEP, LAST_STEP + PERCENTAGE_STEP, PERCENTAGE_STEP):
    ALL_RATE = round(HOUSEHOLD_RATE_100 / (100 - PUBLIC_PERCENTAGE))
    PUBLIC_RATE = ALL_RATE - HOUSEHOLD_RATE
    
    print("공용 사용량 차지비율: {}%\n".format(PUBLIC_PERCENTAGE) +\
         "세대 사용량 (kwh): {}\n".format(HOUSEHOLD_RATE) +\
         "공용 사용량 (kwh): {}\n".format(PUBLIC_RATE) +\
         "전체 사용량 (kwh): {}\n".format(ALL_RATE))
```
